```python
# Use this cell to begin your analysis, and add as many as you would like!
import pandas as pd
logins = pd.read_csv("datasets/logins.csv")
```

```python
valid_passwords = logins[
    (logins['password'].str.len() >= 10) &  # Contraseña de al menos 10 caracteres
    logins['password'].str.contains(r'[a-z]') &  # Contiene al menos una letra minúscula
    logins['password'].str.contains(r'[A-Z]') &  # Contiene al menos una letra mayúscula
    logins['password'].str.contains(r'\d') &  # Contiene al menos un carácter numérico
    logins['password'].str.contains(r'[^\w\s]') &  # Contiene al menos un carácter no alfanumérico
    ~logins['password'].str.contains(r'password', case=False)   # No contiene la palabra "password" (insensible a mayúsculas/minúsculas)
    ]
```

```python
# Crear una lista para almacenar los índices de las filas a eliminar
indices_a_eliminar = []

for index, row in logins.iterrows():
    username = row['username'].lower()
    username = username.replace("."," ")
     
    # Divide el nombre de usuario en primer nombre y apellido
    first_name, last_name = username.split()
    
    password = row['password'].lower()
    # Verifica si el primer nombre o el apellido están en la contraseña
    if (first_name.lower() in password.lower()) or (last_name.lower() in password.lower()):
        indices_a_eliminar.append(index)
    
```

```python
# Filtra el DataFrame para que solo contenga las filas con los índices a eliminar
valid_passwords = valid_passwords.loc[~valid_passwords.index.isin(indices_a_eliminar)]
```

```python
# el numero total menos las pswds validas sobre el total por 100
bad_pass = (len(logins) - len(valid_passwords))/len(logins) * 100
print(bad_pass)
```

```python
# Crear un conjunto de usuarios válidos
usuarios_validos = set(valid_passwords["username"])

# Filtrar los usuarios que no están en el conjunto de usuarios válidos
usuarios_a_cambiar_contraseña = logins[~logins["username"].isin(usuarios_validos)]["username"]

email_list = usuarios_a_cambiar_contraseña.sort_values(ascending = True)
print(email_list)
```
